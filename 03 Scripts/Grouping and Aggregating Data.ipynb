```python
#create new column for avergae prices by user
ords_prods_merge['avg_prices'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
ords_prods_merge.head(15)
```

```python
#creating spender flag to determine types of spenders
ords_prods_merge.loc[ords_prods_merge['avg_prices'] < 10, 'spender_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_prices'] >= 10, 'spender_flag'] = 'High spender'
```

```python
#validating change made above
ords_prods_merge[['user_id', 'spender_flag', 'avg_prices']].tail(60)
```

```python
#counts for each spender type
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
#create new column for median day since prior order by user
ords_prods_merge['median_day'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
ords_prods_merge.head(5)
```

```python
#creating frequency flag to show frequencty of shopping
ords_prods_merge.loc[ords_prods_merge['median_day'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_day'] <= 20) & (ords_prods_merge['median_day'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_day'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
#counts for each customer frequency
ords_prods_merge['frequency_flag'].value_counts(dropna = True)
```
